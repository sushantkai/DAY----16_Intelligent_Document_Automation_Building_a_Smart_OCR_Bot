# A Comprehension of Intelligent Document Processing

Intelligent Document Processing (IDP) is an advanced automation technology that transforms the way businesses handle documents. By leveraging artificial intelligence (AI), IDP systems can automatically read, understand, and process vast quantities of documents, extracting crucial data and feeding it into various business workflows. This technology is a significant leap from traditional Optical Character Recognition (OCR) as it not only digitizes text but also comprehends the context and structure of the information within documents.

At its core, IDP is designed to tackle the challenge of processing unstructured and semi-structured data, which make up a large portion of business documents such as invoices, contracts, emails, and forms. By automating the extraction and interpretation of this data, IDP significantly reduces the need for manual data entry, leading to increased efficiency, higher accuracy, and substantial cost savings.

````python
# Step 4: Automatically detect PDFs in the dataset
pdf_files = []
for root, dirs, files in os.walk("/content/resume_dataset"):
    for file in files:
        if file.lower().endswith(".pdf"):
            pdf_files.append(os.path.join(root, file))

# Take only first 20 PDFs
pdf_files = pdf_files[:20]
print(f"Processing {len(pdf_files)} PDF resumes")
````

````python
# Step 5: Create folder for images
output_image_folder = "/content/resume_images"
os.makedirs(output_image_folder, exist_ok=True)
print(f"Images will be saved to: {output_image_folder}")

# Step 6: Convert PDFs to images
for i, pdf_path in enumerate(pdf_files, 1):
    print(f"Processing PDF {i}/{len(pdf_files)}: {pdf_path}")

    # Convert each page in PDF to an image
    pages = convert_from_path(pdf_path, dpi=200)

    for page_num, page in enumerate(pages, 1):
        image_name = os.path.basename(pdf_path).replace(".pdf", f"_page{page_num}.jpg")
        image_path = os.path.join(output_image_folder, image_name)
        page.save(image_path, "JPEG")
        print(f"Saved image: {image_path}")

print("Conversion of first 20 PDF resumes to images completed.")
````

````python
# Apply Gaussian blur to reduce noise
denoised = cv2.GaussianBlur(gray, (5,5), 0)

# Display denoised image
plt.figure(figsize=(8,10))
plt.imshow(denoised, cmap='gray')
plt.axis('off')
plt.title("Denoised Resume Image")
plt.show()
````

````python
# Apply Otsu's thresholding for binarization
_, binary = cv2.threshold(denoised, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Display binarized image
plt.figure(figsize=(8,10))
plt.imshow(binary, cmap='gray')
plt.axis('off')
plt.title("Binarized Resume Image")
plt.show()
````

````python
import imutils

# Function to deskew an image
def deskew_image(image):
    # If image has 3 channels (color), convert to grayscale
    if len(image.shape) == 3:
        gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        gray_img = image.copy()

    # Find coordinates of all non-zero pixels
    coords = np.column_stack(np.where(gray_img > 0))

    # Compute the angle of the minimum area bounding box
    angle = cv2.minAreaRect(coords)[-1]
    if angle < -45:
        angle = -(90 + angle)
    else:
        angle = -angle

    # Rotate the image to deskew
    deskewed = imutils.rotate_bound(image, angle)
    return deskewed, angle

# Deskew the binarized image
deskewed, angle = deskew_image(binary)

# Display deskewed image
plt.figure(figsize=(8,10))
plt.imshow(deskewed, cmap='gray')
plt.axis('off')
plt.title(f"Deskewed Resume Image (Angle: {angle:.2f})")
plt.show()
````

````python
import cv2
import imutils
import matplotlib.pyplot as plt
import numpy as np
import os

# ---------- Deskew Function ----------
def deskew_image(image):
    if len(image.shape) == 3:
        gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        gray_img = image.copy()

    coords = np.column_stack(np.where(gray_img > 0))
    if len(coords) == 0:
        return image, 0  # No text detected

    angle = cv2.minAreaRect(coords)[-1]
    if angle < -45:
        angle = -(90 + angle)
    else:
        angle = -angle

    if abs(angle) > 45:  # Avoid extreme wrong angles
        angle = 0

    deskewed = imutils.rotate_bound(image, angle)
    return deskewed, angle


# ---------- Paths ----------
image_folder = "/content/resume_images"          # Folder with 20 images
processed_folder = "/content/processed_images"   # Save processed output
os.makedirs(processed_folder, exist_ok=True)

# ---------- Process all images ----------
image_files = sorted(os.listdir(image_folder))[:20]  # Limit to 20

for idx, image_name in enumerate(image_files, 1):
    image_path = os.path.join(image_folder, image_name)
    image = cv2.imread(image_path)

    if image is None:
        print(f" Skipping {image_name} (not readable)")
        continue

    # Step 1: Grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Step 2: Noise Reduction
    denoised = cv2.GaussianBlur(gray, (5, 5), 0)

    # Step 3: Binarization
    _, binary = cv2.threshold(denoised, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Step 4: Deskew
    deskewed, angle = deskew_image(binary)

    # Save processed image
    processed_path = os.path.join(processed_folder, image_name)
    cv2.imwrite(processed_path, deskewed)

    print(f"[{idx}/{len(image_files)}] {image_name}  angle={angle:.2f}, saved at {processed_path}")

# ---------- Display one sample ----------
sample_image = os.path.join(processed_folder, image_files[0])
plt.figure(figsize=(8,10))
plt.imshow(cv2.imread(sample_image, cv2.IMREAD_GRAYSCALE), cmap='gray')
plt.axis('off')
plt.title(f"Sample processed image: {image_files[0]}")
plt.show()
````

Define the Prompt Template

````python
prompt_template = """
Extract information from the given resume text.
Information to extract: name, email, phone, education, skills, experience.
Always respond in this JSON format:
{
    "name": "NAME",
    "email": "EMAIL",
    "phone": "PHONE",
    "education": "EDUCATION",
    "skills": ["SKILL_1", "SKILL_2", "..."],
    "experience": "EXPERIENCE"
}
Correct any obvious OCR errors in the text.

Here is the text:
"""
````

Loop Over OCR Text Files & Extract Information

````python
text_folder = "/content/resume_text"
output_folder = "/content/json_output"
os.makedirs(output_folder, exist_ok=True)

text_files = os.listdir(text_folder)[:20]  # Only first 20 resumes
start_time = time.time()

for i, txt_file in enumerate(text_files, 1):
    text_path = os.path.join(text_folder, txt_file)

    with open(text_path, "r") as f:
        text = f.read()

    # Create the prompt for this resume
    prompt = prompt_template + text

    # Prepare contents (image optional, can add later)
    contents = [
        {
            "text": prompt
        }
    ]

    # Call GenAI model
    response = genai_client.models.generate_content(
        model='gemini-2.5-flash',
        contents=contents
    )

    # Parse output JSON
    extracted_info = json.loads(response.text.replace('```json','').replace('```',''))

    # Save JSON
    output_path = os.path.join(output_folder, txt_file.replace(".txt", ".json"))
    with open(output_path, "w") as f:
        json.dump(extracted_info, f, indent=4)

    print(f"[{i}/{len(text_files)}] Extracted info saved: {output_path}")
    time.sleep(5)  # Optional: avoid rate limits

print(f"Information extraction completed in {time.time() - start_time:.2f} seconds.")
````
